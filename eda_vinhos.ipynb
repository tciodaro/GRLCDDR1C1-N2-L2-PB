## Visualizar Matriz de CorrelaÃ§Ã£o

### Subtask:
Gerar um mapa de calor (heatmap) da matriz de correlaÃ§Ã£o das variÃ¡veis em `feature_list` do DataFrame `wine` para visualizar as relaÃ§Ãµes entre elas.


**Reasoning**:
I will import the necessary libraries, calculate the correlation matrix for the specified features, and then generate a heatmap to visualize these correlations with annotations and a coolwarm colormap, adding a title for clarity.



```python
import seaborn as sns
import matplotlib.pyplot as plt

correlation_matrix = wine[feature_list].corr()

sns.clustermap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f", figsize=(12, 10))
plt.suptitle('Matriz de CorrelaÃ§Ã£o das VariÃ¡veis do Vinho (Clustered)', y=1.02)
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

# Melt the DataFrame to have 'variable' and 'value' columns for easier plotting
wine_melted = wine[feature_list].melt(var_name='VariÃ¡vel', value_name='Valor')

plt.figure(figsize=(16, 8)) # Adjust figure size for better readability

sns.boxplot(x='VariÃ¡vel', y='Valor', data=wine_melted, palette='viridis')
plt.title('Box Plots de Todas as VariÃ¡veis do Vinho', fontsize=16)
plt.xlabel('VariÃ¡vel', fontsize=12)
plt.ylabel('Valor', fontsize=12)
plt.xticks(rotation=45, ha='right') # Rotate x-axis labels for better readability
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout() # Adjust layout to prevent overlapping labels
plt.show()
```
